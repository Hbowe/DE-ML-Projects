```python
## Functions for DF cleaning ##

## Adjusting column names of semesters to match the more common name ##
fl2020 = fl2020.rename(columns={'Catalog Nbr': 'Catalog Number'})
sp2021 = fl2021.rename(columns={'Catalog Nbr': 'Catalog Number'})

## Condensing repititive DF filtering ##

def oneclass(semester, subject, elective):
    electives = (semester["Subject"] == subject) & (semester["Catalog Number"] == elective)
    return electives

def multiclass(semester, subject, electivelst):
    electives = (semester["Subject"] == subject) & (semester["Catalog Number"].isin(electivelst))
    return electives

def upperclass(semester, subject, elective):
    upperrange = (semester["Subject"] == subject) & (semester["Catalog Number"] > elective)
    return upperrange

def dropclass(semester, subject, elective):
    dropclass = semester[(semester["Subject"] == subject) & (semester["Catalog Number"] == elective)].index
    dropclass = semester.drop(dropclass, inplace = True)
    return dropclass

def dropclasses(semester, subject, electivelst):
    dropclasses = semester[(semester["Subject"] == subject) & (semester["Catalog Number"].isin(electivelst))].index
    dropclasses = semester.drop(dropclasses, inplace = True)
    return dropclasses

def droprange(semester, subject, elective):
    dropclasses = semester[(semester["Subject"] == subject) & (semester["Catalog Number"] < elective)].index
    dropclasses = semester.drop(dropclasses, inplace = True)
    return dropclasses

```

```python
## Merging DFs ##

## Baked in the removal of columns like Incompletes, W drops, Pass/Fails and similar not GPA producing course columns since their outcomes are typically related to external factors and are not usually direct indicators of course quality/rigor ##
## if you do add back columns to weigh external forces, be aware that not all CSVs have I column and may lead to rows being chopped off ##
## I have left The instructor columns alone because most of the time only the first 2 are used, but randomly 1 class will use all 6 and some sample sizes are very small for some majors where losing 1 class with the use of drop.na could be impactfull ##

def mergesem(sem1, sem2, sem3, sem4, sem5):
    lst = [sem1, sem2, sem3, sem4, sem5]
    for i in lst:
        for column in i:
            if column == "Section":
                i.drop(column, axis = 1, inplace=True)
            if column == "P":
                i.drop(column, axis = 1, inplace=True)
            if column == "W":
                i.drop(column, axis = 1, inplace=True)
            if column == "NC":
                i.drop(column, axis = 1, inplace=True)
            if column == "CR":
                i.drop(column, axis = 1, inplace=True)
            if column == "I":
                i.drop(column, axis = 1, inplace=True)
    merge = pd.concat(lst, ignore_index=True)
    return merge

ACCTfull = mergesem(ACCTsp2020, ACCTfl2020, ACCTsp2021, ACCTfl2021, ACCTsp2022)## Accounting requires 12 elective hours ##
BABEfull = mergesem(BABEsp2020, BABEfl2020, BABEsp2021, BABEfl2021, BABEsp2022)## BA Business Economics Con - 9 hours ##
BAEMfull = mergesem(BAEMsp2020, BAEMfl2020, BAEMsp2021, BAEMfl2021, BAEMsp2022)## BA Energy Management Con - 12 hours ##
BAIEfull = mergesem(BAIEsp2020, BAIEfl2020, BAIEsp2021, BAIEfl2021, BAIEsp2022)## BA Innovation and Entrepreneurship Con - 12 hours ##
BARMfull = mergesem(BARMsp2020, BARMfl2020, BARMsp2021, BARMfl2021, BARMsp2022)## BA Risk Management and Insurance Con - 9 hours ##
BASfull = mergesem(BASsp2020, BASfl2020, BASsp2021, BASfl2021, BASsp2022)## BA Sales Con - 12 hours ##
BAGBfull = mergesem(BAGBsp2020, BAGBfl2020, BAGBsp2021, BAGBfl2021, BAGBsp2022)## BA General Business Option - 15 hours ##
BAREfull = mergesem(BAREsp2020, BAREfl2020, BAREsp2021, BAREfl2021, BAREsp2022)## BA Real Estate Investment Management Con - 12 hours (this concentration allows for any JSOM upp division course as electives + 2 singled out specific entries) ##
BAfull = mergesem(BAsp2020, BAfl2020, BAsp2021, BAfl2021, BAsp2022)## Business Analystics - 9 hours ##
FIfull = mergesem(FIsp2020, FIfl2020, FIsp2021, FIfl2021, FIsp2022)## Finance - 21 hours ##
FIEfull = mergesem(FIEsp2020, FIEfl2020, FIEsp2021, FIEfl2021, FIEsp2022)## Finance and Economics double major ##
GBECfull = mergesem(GBECsp2020, GBECfl2020, GBECsp2021, GBECfl2021, GBECsp2022)## GB Business Economics Concentration - 15 hours ##
GBEMfull = mergesem(GBEMsp2020, GBEMfl2020, GBEMsp2021, GBEMfl2021, GBEMsp2022)## GB Energy Management Con - 12 hours ##
GBIEfull = mergesem(GBIEsp2020, GBIEfl2020, GBIEsp2021, GBIEfl2021, GBIEsp2022)## GB Innovation and Entrepreneurship Con - 12 hours##
GBRMfull = mergesem(GBRMsp2020, GBRMfl2020, GBRMsp2021, GBRMfl2021, GBRMsp2022)## GB Risk Management and Insurance Con - 15 hours ##
GBREfull = mergesem(GBREsp2020, GBREfl2020, GBREsp2021, GBREfl2021, GBREsp2022)## GB Real Estate Investment Management Con - 12 hours ##
GBSfull = mergesem(GBSsp2020, GBSfl2020, GBSsp2021, GBSfl2021, GBSsp2022)## GB Sales Con ##
GBHRfull = mergesem(GBHRsp2020, GBHRfl2020, GBHRsp2021, GBHRfl2021, GBHRsp2022)## Global Business and Human Resource Management double major  - 5 hours ##
GBMfull = mergesem(GBMsp2020, GBMfl2020, GBMsp2021, GBMfl2021, GBMsp2022)## Global Business and Maketing double major - 5 hours ##
GBSCfull = mergesem(GBSCsp2020, GBSCfl2020, GBSCsp2021, GBSCfl2021, GBSCsp2022)## Global Business and Supply Chain Managment double major - 5-8 hours ##
HMfull = mergesem(HMsp2020, HMfl2020, HMsp2021, HMfl2021, HMsp2022)## Healthcare Management - 12 hours ##
HMBfull = mergesem(HMBsp2020, HMBfl2020, HMBsp2021, HMBfl2021, HMBsp2022)## Healthcare Management and Biology double major - 21 hours ##
HMMBfull = mergesem(HMMBsp2020, HMMBfl2020, HMMBsp2021, HMMBfl2021, HMMBsp2022)## Healthcare Management and Molecular Biology double major - 21 hours ##
HRMfull = mergesem(HRMsp2020, HRMfl2020, HRMsp2021, HRMfl2021, HRMsp2022)## Human Resource Management - 9 hours ##
ITSSfull = mergesem(ITSSsp2020, ITSSfl2020, ITSSsp2021, ITSSfl2021, ITSSsp2022)## Information Technology and Systems - 12 hours ##
MARfull = mergesem(MARsp2020, MARfl2020, MARsp2021, MARfl2021, MARsp2022)## Marketing - 12 hours ##
SCMfull = mergesem(SCMsp2020, SCMfl2020, SCMsp2021, SCMfl2021, SCMsp2022)## Supply Chain Management - 12 hours ##
```
